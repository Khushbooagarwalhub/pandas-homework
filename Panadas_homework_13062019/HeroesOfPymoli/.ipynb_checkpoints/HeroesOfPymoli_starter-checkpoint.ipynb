```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

# Player Count

* Display the total number of players


```python

#calculate the count of unique player
total_players = purchase_data["SN"].value_counts()
#total_players
#calculate the total number of players
total_players_count = len(total_players)
#place the number of players in a dataframe
total_players_df = pd.DataFrame({"Total players":[total_players_count]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate number of unique items
items_count = len(purchase_data["Item ID"].unique())
items_count
#calculate the average price using mean function
average_price = purchase_data["Price"].mean()
average_price
#calculate the total revenue 
total_revenue = purchase_data["Price"].sum()
total_revenue
#Find number of purchases
purchase_number = purchase_data["Purchase ID"].count()
purchase_number

#create a summary table and store the above data in it
summary_table1 = pd.DataFrame({"Number of Unique Items":[items_count],
                               "Average Price":[average_price],
                               "Number of Purchases":[purchase_number],
                               "Total Revenue":[total_revenue]})
#round the Average price and Total Revenue to nearest cent, include a dollar sign, and use comma notation
summary_table1["Average Price"] = summary_table1["Average Price"].apply("${:,.2f}".format)
summary_table1["Total Revenue"] = summary_table1["Total Revenue"].apply("${:,.2f}".format)

summary_table1
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

#create bins for age 
bins = [0, 9, 14, 19, 24, 29, 34, 39,purchase_data["Age"].max()]
#assign group names to the bins
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#use pd.cut to assign the corresponding bins in the table
purchase_data["labels_age"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
# group the table based on age bins 
age_grouped_data = purchase_data.groupby("labels_age")
age_grouped_data.count().head()

#calculate the total count for the grouped data
total_count = age_grouped_data["SN"].nunique()
total_count
# calculate the percentage of players for the grouped data
percentage = (total_count/total_players_count)*100
percentage

#create a summary table to store the above values
summary_table_4 = pd.DataFrame({"Total Count":total_count,
                                "Percentage of players":percentage})

#round the percentage  to 2 decimal places and include a percentage  sign
summary_table_4["Percentage of players"] = summary_table_4["Percentage of players"].apply('{:.2f}%'.format)
summary_table_4



```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create a new dataframe and store Item ID,Item Name and Item Price
new_df = purchase_data[["Item ID","Item Name","Price"]]
new_df
#group by the Item ID and Item Name
grouped_id_name = new_df.groupby(["Item ID","Item Name"])
grouped_id_name.count().head()

#calculate the purchase count for the grouped data 
new_purchase_count = grouped_id_name["Item ID"].count()
new_purchase_count
#calculate the average price for the grouped data
item_price = grouped_id_name["Price"].mean()
item_price
#calculate the total price for the grouped data
new_total_purchase = grouped_id_name["Price"].sum()
new_total_purchase

#create a summary table and store the values above
summary_table_7 = pd.DataFrame({"Purchase Count":new_purchase_count,
                               "Item Price":item_price,
                               "Total Purchase value":new_total_purchase})
                              
summary_table_7

#round the Item price to nearest cent, include a dollar sign, and use comma notation
summary_table_7["Item Price"] = summary_table_7["Item Price"].apply("${:,.2f}".format)

#create a temporary series to store total purchase value before formatting it
temp_table_series = summary_table_7["Total Purchase value"]

#round the total purchase price to nearest cent, include a dollar sign, and use comma notation
summary_table_7["Total Purchase value"] = summary_table_7["Total Purchase value"].apply("${:,.2f}".format)
summary_table_7

#sort the table on basis of Purchase count
summary_table_7 = summary_table_7.sort_values("Purchase Count",ascending=False)
summary_table_7.head()


```
